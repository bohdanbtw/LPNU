```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np



df = pd.read_csv('run_gyroscope.csv', index_col=0)
df = df.rename(columns={'z': 'z_gyroscope', 'y': 'y_gyroscope', 'x': 'x_gyroscope'})

df1 = pd.read_csv('run_accelerometer.csv', index_col=0)
df1 = df1.rename(columns={'z': 'z_accelerometer', 'y': 'y_accelerometer', 'x': 'x_accelerometer'})

df1 = df1.drop(['seconds_elapsed'], axis=1)
df2 = df.join(df1)

df21 = pd.DataFrame()

# Calculate sums for each column
for col in df2.columns:
    # Calculate sum of element i and element i+2
    sum_col = df2[col] - df2[col].shift(-4)
    # Store the sums in the new DataFrame
    df21[f'{col}'] = sum_col

# Drop NaN values at the end
df21.dropna(inplace=True)
```

```python
df21["Class"] = 2
df21.head()
```

```python

df = pd.read_csv('walk_gyroscope.csv', index_col=0)
df = df.rename(columns={'z': 'z_gyroscope', 'y': 'y_gyroscope', 'x': 'x_gyroscope'})

df1 = pd.read_csv('walk_accelerometer.csv', index_col=0)
df1 = df1.rename(columns={'z': 'z_accelerometer', 'y': 'y_accelerometer', 'x': 'x_accelerometer'})

df1 = df1.drop(['seconds_elapsed'], axis=1)
df3 = df.join(df1)


df31 = pd.DataFrame()

# Calculate sums for each column
for col in df3.columns:
    # Calculate sum of element i and element i+2
    sum_col = df3[col] - df3[col].shift(-4)
    # Store the sums in the new DataFrame
    df31[f'{col}'] = sum_col

# Drop NaN values at the end
df31.dropna(inplace=True)


df31["Class"] = 0
df31.head()
```

```python

df = pd.read_csv('pres_gyroscope.csv', index_col=0)
df = df.rename(columns={'z': 'z_gyroscope', 'y': 'y_gyroscope', 'x': 'x_gyroscope'})

df1 = pd.read_csv('pres_accelerometer.csv', index_col=0)
df1 = df1.rename(columns={'z': 'z_accelerometer', 'y': 'y_accelerometer', 'x': 'x_accelerometer'})

df1 = df1.drop(['seconds_elapsed'], axis=1)
df4 = df.join(df1)


df41 = pd.DataFrame()

# Calculate sums for each column
for col in df4.columns:
    # Calculate sum of element i and element i+2
    sum_col = df4[col] - df4[col].shift(-4)
    # Store the sums in the new DataFrame
    df41[f'{col}'] = sum_col

# Drop NaN values at the end
df41.dropna(inplace=True)


df41["Class"] = 1
df41.head()
```

```python
dataset = pd.concat([df41,df21,df31], ignore_index=True)

dataset.head()
```

```python
dataset = dataset.drop(["seconds_elapsed"], axis = 1)
dataset.tail()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Поділ даних на тренувальну і валідаційну вибірки (70/30%)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Нормалізація даних
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_val_normalized = scaler.transform(X_val)


# Побудова кореляційної матриці
correlation_matrix = pd.DataFrame(X_train_normalized).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score, confusion_matrix

# Модель
model = nn.Sequential(
    nn.Linear(6, 128),
    nn.ReLU(),
    nn.Linear(128, 22599),
    nn.ReLU(),
    nn.Linear(22599, 3)
)

# Визначення функції втрат та оптимізатора
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Генерація випадкових даних для тренування
X_train_tensor = torch.FloatTensor(X_train_normalized)
y_train_tensor = torch.LongTensor(y_train)

# Навчання моделі
num_epochs = 115
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    # Backward pass та оптимізація
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 1 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Перевірка точності моделі на валідаційних даних
# Переконайтеся, що X_val_normalized має форму (N, 64)
X_val_tensor = torch.FloatTensor(X_val_normalized)
y_val_tensor = torch.LongTensor(y_val)

# Перевірка розміру вхідного тензора
print(f"Shape of validation input tensor: {X_val_tensor.shape}")

with torch.no_grad():
    outputs = model(X_val_tensor)
    _, predicted = torch.max(outputs.data, 1)
    accuracy = accuracy_score(y_val_tensor.numpy(), predicted.numpy())
    print(f'Accuracy on the validation set: {accuracy:.2f}')

# Обчислення матриці конфузії
cm = confusion_matrix(y_val_tensor.numpy(), predicted.numpy())
print("Confusion Matrix:")
print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
# Обчислення матриці конфузії


# Функція для побудови матриці конфузії
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='.0f', cmap=cmap, xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

# Виклик функції для побудови графічного представлення матриці конфузії
plot_confusion_matrix(cm, [0,1,2])
```
